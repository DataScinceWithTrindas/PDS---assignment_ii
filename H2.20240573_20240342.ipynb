```python
# Filter countries with a population between 9 and 11 million
population_filter = data[(data["Population"] >= 9000000) & (data["Population"] <= 11000000)]

# Display the result
print(population_filter[["Country Code", "Population"]])
```

```python
#2
# Calculate the average and standard deviation of GDP for the filtered countries
average_gdp = population_filter["GDP"].mean()
std_gdp = population_filter["GDP"].std()

# Display result
print(f"Average GDP (in $): {average_gdp}, Standard Deviation of GDP (in $): {std_gdp}")
```

```python
#3 Chatgpt to see the formulas
# Make a copy of the filtered DataFrame to avoid SettingWithCopyWarning
population_filter_copy = population_filter.copy()

# GDP per capita 
population_filter_copy.loc[:, "GDP per Capita"] = population_filter_copy["GDP"] / population_filter_copy["Population"]

# average and standard deviation of GDP per capita
average_gdp_per_capita = population_filter_copy["GDP per Capita"].mean()
std_gdp_per_capita = population_filter_copy["GDP per Capita"].std()

# Display the result
print(f"Average GDP per Capita: {average_gdp_per_capita}, Standard Deviation of GDP per Capita: {std_gdp_per_capita}")
```

```python
#4 didnt know if professor want to use all data or the countries from previous code. I used all data
# function to classify country size based on population
def classify_country_size(population):
    if population < 1_000_000:
        return "Tiny"
    elif 1_000_000 <= population < 5_000_000:
        return "Very Small"
    elif 5_000_000 <= population < 15_000_000:
        return "Small"
    elif 15_000_000 <= population < 30_000_000:
        return "Medium"
    elif 30_000_000 <= population < 100_000_000:
        return "Large"
    else:
        return "Huge"

# do the function to create a new column in the data DataFrame
data["Size Classification"] = data["Population"].apply(classify_country_size)

# Ensure GDP per capita is calculated correctly in the data DataFrame
data["GDP per Capita"] = data["GDP"] / data["Population"]

# Group by the size classification and calculate the average and standard deviation of GDP per capita
grouped_stats = data.groupby("Size Classification")["GDP per Capita"].agg(["mean", "std"])

# Display the result
print(grouped_stats)
```

```python
#5 A little prompt engineering was used to complete this exercise
def above_mean(df, column_name):
    # we calculate the mean of the column, excluding NaN values
    column_mean = df[column_name].mean()

    # function to calculate if above or below
    def check_above_mean(value):
        if pd.isna(value):
            return np.nan  # Keep NaN values as NaN
        return 1 if value > column_mean else 0

    # do the function to the specified column
    return df[column_name].apply(check_above_mean)

# Test the function on the 'GDP per Capita' column in the 'data' DataFrame
result = above_mean(data, "GDP per Capita")
print(result.head())
```

```python
#6
# mean human capital value
human_capital_mean = data["hc"].mean()

# column to indicate whether human capital is above or below the average
data["Human Capital Above Avg"] = data["hc"].apply(lambda x: "Above" if x > human_capital_mean else "Below")

# group by size classification and whether human capital is above or below the mean
gdp_by_size_hc = data.groupby(["Size Classification", "Human Capital Above Avg"])["GDP per Capita"].mean()

# display the result
gdp_by_size_hc.head()
```

```python
#7
# mean gini coefficient
gini_mean = data["GINI"].mean()

# column to indicate whether the gini coefficient is above or below the mean
data["Gini Above Avg"] = data["GINI"].apply(lambda x: "Above" if x > gini_mean else "Below")

# group by human capital and gini coefficient above/below average
gdp_by_hc_gini = data.groupby(["Human Capital Above Avg", "Gini Above Avg"])["GDP per Capita"].mean()

# Display the result
print(gdp_by_hc_gini)
```

```python
#8 
# Filter countries with GINI and human capital below average
filtered_data = data[(data["GINI"] < gini_mean) & (data["hc"] < human_capital_mean)]

# Find the country with the highest GDP per Capita
highest_gdp_country = filtered_data.loc[filtered_data["GDP per Capita"].idxmax()]["Country Code"]

# Display the result
print("Country with highest GDP per capita, Gini below average, and human capital below average:", highest_gdp_country)
```

```python
#9 this is similar to the previous question, but now also group by size classification.
# Create a DataFrame with countries that have GINI and human capital below average for their size classification
filtered_data = data[(data["GINI"] < gini_mean) & (data["hc"] < human_capital_mean)]

# Find the country with the highest GDP per capita in this filtered data
highest_gdp_country = filtered_data.loc[filtered_data["GDP per Capita"].idxmax()]["Country Code"]

# Display the result
print("Country with highest GDP per capita (size classification, GINI below avg, Human Capital below avg):", highest_gdp_country)
```
